# Part 1

```python
def load_data(file:str):
    raw = open(file, 'r').read()
    raw = raw.split(',')
    raw = [int(x) for x in raw]
    return raw
test_data = load_data('day_7_input_test.txt')
data = load_data('day_7_input.txt')
```

```python
tmp = data
def cum_sum(x_i, fixed_x_j):
    j=0
    for i in range(x_i,fixed_x_j):
        j+=i
    return j
    # return np.sum(list(range(x_i,fixed_x_j)))
```

```python

max_position = np.max(tmp) + 1
min_position = np.min(tmp)
fuel = []
for i in range(min_position, max_position):
    current_fuel = 0
    for c in tmp:  
        current_fuel += np.abs(c-i)
    fuel.append(current_fuel)
```

```python
np.min(fuel)
```

```python

max_position = np.max(tmp) + 1
min_position = np.min(tmp)

fuel = []
for i in tqdm(range(min_position, max_position)):
    current_fuel = 0
    for c in tmp:  
        dif = np.abs(c-i) + 1
        current_fuel += cum_sum(1,dif)
    fuel.append(current_fuel)
print(fuel)
```

```python
np.min(fuel)
```
